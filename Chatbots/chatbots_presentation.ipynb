## Úvod

### Co je to chatbot?

Na nejnižší úrovni vnímáme chatboty jako formu nového uživatelského rozhraní. Toto nové oživatelské rozhraní umožňuje uživatelům interaktivně využívat služby a brand prostřednictvím konverzačních platforem [1].

### Conversational AI

Technologie, která zabezpečuje automatizované konverzace mezi byznysem a zákazníky. Konverzace mohou probíhat **textově**, **hlasově formou zpráv**, nebo na **voice-based komunikační platformě**.

AI se vyvinula na takovou úroveň, že dokáže simulovat lidskou konverzaci s vysokou přesností. Klíčem k dosažení tak vysoké úrovně lidské konverzace je využití **NLP (Natural Language Processing)**. NLP pomáchá chatbotům porozumět **lidské řeči**, **textu**, **záměru**, **rozpoznat jazyk**, **porozumět kontextu**, či **personalizovat konverzaci**, a nakonec **reagovat jako člověk** [2].

## Historie chatbotů

### Slovo "robot"

**Chatbot = chat + bot**. Slovo robot se dostalu do slovníků světových jazyků díky českému spisovateli **Karlovi Čapkovi** a jeho sci-fi dílu **R.U.R (Rossumovi univerzální roboti)**.

### ELIZA, rok 1960

*Chatbot: https://www.masswerk.at/elizabot/*

Eliza přijímá a zkoumá na vstupu klíčová slova, která analyzuje pomocí definovaného souboru pravidel a generuje výstup. Ve svém výstupu použije buď klíčové slovo se vstupu a nebo použije generickou odpověď. Autorem tohoto NLP programu je Joseph Weizenbaum, který na něm demonstroval dialog mezi člověkem a počítačovým programem na bázi Rogersova přístupu (v překladu přístup zaměřený na člověka). Eliza dosahuje nejlepších terapeutických výsledků když je uživatel vyzván k popisování jeho pocitů. Takový způsob rozhovoru byl zvolený z toho důvodu, že psychiatrický rozhovor je jedním z mála příkladů kategorizované dialogové komunikace v přirozeném jazyce, ve kterém jeden ze zůčastněných může předstírat, že neví téměř nic o skutečném světě.

![Eliza](img/eliza.png)

### PARRY, rok 1972, autor Kenneth Colby

Kenneth Colby vytvořil chatbota, který modeloval chování paranoidního schizofrenického pacienta. Podle Kennetha Colbyho byl **PARRY** něco jako **Eliza** s určitým postojem [4].

V lednu roku 1973 podstoupil průkopník počítačové vědy Vint Cerf během počítačové mezinárodní konference ony dva zmíněné chatboty (Elisa vs PARRY) vzájemné konverzaci. Tuto událost označil jako vzájemné střetnutí počítačových myslí.

![Eliza vs PARRY](img/eliza_vs_parry.png)

*Ukázka výsledku této konverzace: https://tools.ietf.org/html/rfc439*

### Eugene Goostman, rok 2014

*Chatbot: http://eugenegoostman.elasticbeanstalk.com*

Chatbot, který v roce 2014 přesvědčil přesvědčil nic netušící porotce soutěžní hry *Can Machines Think?*, že je člověk [5]. Eugene Goostman prošel Turingovým testem, který zkoumá, zda lidští tazatelé odhhalí respondenta, který není člověkem. Soutěž se konala při příležitosti 60. výročí Turingovi smrti. V sérii 5minutévých rozhovorů se podařilo Eugenovi přesvědčit 33 % rozhodčí, že je člověk, což ještě nikdy předtím žádný jiný program nedokázal.

## Typy chatbotů

Amir Shewat [1] uvádí ve své knize následující typy chatbotů.

### Personal vs Team Bots

*Personal bot* (osobní / soukromý bot) slouží jako osobní asistent. Tento bot je zaměřen na komunikaci s jedním uživatelem. Příkladem osobního chatbota může být např. nákupní bot na **Facebook Messengeru**, inteligentní agent **J.A.R.V.I.S** ve filmu **Iron Man**, či chatbot ve filmu **Her**.

*Team bot* (týmový bot) usnadňuje týmové procesy a aktivity, např. bot *Lunch Train* pomáhá si týmům vybrat, kam jít společně na obědm či *Standup Bot*, který usnadňuje týmové standupy. Týmový bot může může s více uživateli komunikovat buď soukromě, nebo veřejně ve společném kanálu.

Týmové chatboty bývají většinou komplexnější, zatímco osobní boty mohou být na implementaci jednodušší, avšak pokrývají menší množinu use cases. Osobní chatboty mohou umožňovat komunikaci s více osobami, ale vždy mohou v jednom okamžiku v daném kontextu komunikovat pouze s jedním člověkem.

### Super Bots vs Domain-Specific Bots

*Domain-specific bot* (donénově specifický bot) je vytvořený pro nějakou konkrétní službu (produkt, nebo brand). Takový chatbot představuje tuto službu a uživatelé si chatbota s touto službou asociují. Příkladem může být chatbot letecké společnosti, který dokáže klientům pomoci se vším od bookování letů až po upozornění na události ohledně daného letu.

*Super bot* je druh chatbota, který dokáže poskytovat více různých služeb a mohou také umožňovat použití doplňkových modulů pro rozšíření jejich funkcionality. Typickým příkladem tohoto typu chatbota je *Google Assistant*, jenž poskytuje různé Google služby, jako jsou např. mapy, počasí, hry, ale také umožňuje např. poskytovat data z Wikipedie.

Zatímco výhodou doménově specifického bota je, že dokáže poskytnout uživatelům v konzistentní formě informace z mnoha oblastní, tak výhotou super bota je možnost specializace pro specifický obsah a účel. Při vytváření super botů vedou snahy o co možná největší standardizaci a při integraci vlastní služby máme menší kontrolu nad tím, jak bude chatbot naši službu prezentovat. V případě vytváření doménově specifického bota nad nám máme větší kontrolu a jsme tak schopni více ovlivnit *user experience*.

### Business Bots vs Consumer Bots

*Business bot* a *consumer bot* se odlišují v mnoha ohledech. Zatímco účelem business botů je podpora nějakých úkolů, či business procesů, se zaměřením na produktivitu a věcnost, tak consumer bot může být více "upovídaný" a v jeho projevu může být patrné emotivní zabarvení, např. může používat humor.

### Voice vs Text Bots

Dalším parametrem, podle kterého můžeme chatboty dělit, je způsob konverzace. V dnešní době se můžeme běžně setkat s boty, které dvě hlavní možnosti komunikace - hlasovou a textovou. Zástupce chatbotů, které komunikují v textové formě, nalezneme nejčastněji v chatovacích aplikacích, jako jsou např. **Slack**, **Facebook**, či **Telegram**, avšak mohou disponovat vlastní vyhrazenou aplikací. Mezi nejznámnější zástupce hlasových botů řadíme **Alexu** od Amazonu, **Cortanu** od Microsoftu, **Siri** od Applu, **Google Assistant** a mnoho dalších. Konverzace s těmito chatboty je většinou iniciována nějakým hlasovým povelem, či stisknutím tlačítka a konverzace probíhá ve formě krátkých příkazů, nebo otázek / odpovědí.

## AI chatbotů

### Natural Language Understanding (NLU)

NLU umožňuje chatbotům odvozovat **intent** (záměr) z přirozeného jazyka [1]. Tento proces většinou zahrnuje strojové učení za využití velkého datasetu vzorových konverzací. Většina NLU frameworků nám umožňuje překlad ze vstupu zadaného uživatelem, jako je např. věta *"Chci si koupit lístek do divadla"*, na nějaký *intent*, v tomto případě koupit lístek. Tyto frameworky jsou schopny rovněž extrahovat **entity**, což jsou proměnné označující název určitého objektu, či nějakou hodnotu, např. jméno, datum, adresa, číslo, atd.

### Conversational  Management

Extrakce intentů a entit je důležitým, avšak pouze základním, krokem pro řízení konverzace [1]. Další vrstou je *conversational management* řídící samotné toky konverzace. Tato vrstva využívá velmi vyspělé prvky umělé inteligence umožňující porozumět kontextu konverzace a ví jak se navigovat mezi různými kontexty a "podkonverzacemi" (tyto konverzační toky označujeme někdy jako *stories*). Extrakce záměru je v komplexních konverzacích velmi složitá, jelikož lidé během konverzace snadno odběhnou k jinému tématu a bez umělé inteligence by bylo extrémně obtížné extrahovat správný záměr a vést pomocí chatbota složité konverzace.

### Image Recognition / Computer Vision

Pomocí metod zpracování obrazu mohou dnes chatboty např. rozpoznávat obrázky, extrahovat z obrázků text, či odhadovat emoce.

### Prediction

API poskytující predikci nejsou specifická pouze pro oblast chatbotů, ale mohou být využita pro optimalizaci konverzací. Díky predikci se mohou chatboty na historických datech naučit, jakou odpověď bude v konrkétním případě nejlepší zvolit a tím dokážou i vhodněji cílit na zákazníka.

### Sentiment Analysis

Jedním z nástrojů umožňujících emoční analýzu konverzace je *IBM Watson*. Tyto nástroje umožňují rozpoznat aktuální emoční stav člověka a podle něj přizpůsobit způsob další komunikace. Správně strojově rozpoznat z konverzace emoce je poměrně náročný úkol a prozatím je tato oblast spíše na počátku svého vývoje a neposkytuje příliš přesné výsledky.

### Wit.ai
*https://wit.ai*

```python
from wit import Wit
import datetime

client = Wit('75IDAKN7KZJGQQZ4TI54YVPYUDFMV275')

all_simple_intents = {
    'greeting': 'Ahoj. Jsem VuiBot a rád ti zodpovím tvoje dotazy ohledně předmětu umělá inteligence 2.',
    'requirements': 'Pro ukončení předmětu je potřeba odprezentovat zvolené téma.',
    'who_teach': 'Předmět umělá inteligentce 2 (VUI2) vyučuje Ing. Jan Kolomazník, Ph.D.'
}

topic_date = {
    'hraní her' : datetime.datetime(2020, 11, 3),
    'clusterizace' : datetime.datetime(2020, 11, 3),
    'zpracování obrazu' : datetime.datetime(2020, 11, 10),
    'stavový prostor' : datetime.datetime(2020, 11, 10),
    'regrese' : datetime.datetime(2020, 11, 17),
    'chatbot' : datetime.datetime(2020, 11, 24),
    'klasifikace' : datetime.datetime(2020, 12, 1),
    'zpracování řeči' : datetime.datetime(2020, 12, 8),
}

def send_message(question):
    print('VuiBot: ' + handle_message(client.message(question)))    

def first_value(obj, key):
    if not obj:
        return None
    if key not in obj:
        return None
    val = obj[key][0]['value']
    if not val:
        return None
    return val

def handle_message(response):
    topic = first_value(response['entities'], 'topic:topic')
    if response['intents']:
        first_intent = response['intents'][0]
        if first_intent['name'] in all_simple_intents:
            return all_simple_intents[first_intent['name']]
        elif topic is not None:
            return 'Prezentace na téma ' + topic + ' se bude konat ' + topic_date[topic].strftime("%d. %m. %Y") + '.'
        elif topic is None:
            return 'Prezentace na zadané téma není naplánovaná.'
    else:
        return 'Můžu ti oskytnout informace o předmětu VUI2. Zeptej se mě na něco ;)'

while True:
        question = input('Me: ')
        send_message(question)
```

### Rasa

*https://rasa.com*

Vytváranie kontextových asistentov a chatbotov. Rasa poskytuje nástroje potrebné pre vysoko výkonných kontextových asistentov. Tvorba kontextového asistenta vyžaduje machine learning. Okrem odpovedania na FAQ a vykonávania úloh, sa taký kontextový asistent od základného chatbota odlišuje najmä v tom, že pokračuje flexibilne  v konverzácii s ohladom na spätný a možný prichádzajúci kontext.

**Čo je to kontext?**

Podľa výkladu Rasa je to schopnosť porozumieť relevancii predcházajúcich správ a dosiahnutiu cieľa užívateľa čo najefektívnejšie. Pochopenie kontextu je kľúčovým krokom k vykonaniu správnej akcie závislej od situácie. 

Konverzačná AI aplikácia sa vytvorená z niekoľkých vrstiev:

![Chatbot UI](img/conversational_layers.png)
(Rasa, 2020)

- Computation Layer: tvoria ju tzv. machine learning-friendly programovacie jazyky ako napr. Python, a taktiež frameworky ako TensorFlow, NumPy a pod.

- Conversational AI Infrastructure Layer: Rasa Open Source ako štandard s building blocks potrebnými pre implementáciu asistenta, pri ktorom nemusíme programovať tzv od základov.

- Tool Layer: vrstva potrebná pre testovanie a zlepšovanie asistenta. Rasa X má sadu nástrojov pre zbieranie, vizualizovanie a náhľad uživateľských konverzácií. 

- Application Layer: prostredie na interakciu s uživateľom.


**Ako funguje RASA?**

Open-source machine learning framework automatizuje textové a hlasové konverzácie. Má základné dve komponenty:

- Rasa NLU: vykonáva základné NLU úlohy, napr. klasifikácia zámeru (aký požiadavok?) a rozpoznanie entity (predmet požiadavku, kľúčová informácie) z uživateľského inputu. Cieľom je teda extrakcia štrukturovanej informácie zo správy. 

- Rasa Core: navrhuje dizajn konverzácie, jej priebeh, a akcie založené na predošlých vstupov.

- NLU: intent klasifikace a entity extraction
- Dialogue management: odpovedanie uživateľovi správou alebo volaním API.
